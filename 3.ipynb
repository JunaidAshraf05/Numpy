```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print(" ====  sales analysis  ====  ")
print("\n sales data shape: ", sales_data.shape)
print("\n sales data: \n", sales_data)
print("\n  1st 3 rows sales data : ", sales_data[0:3])


```

```python
#total sales per year
print(np.sum(sales_data, axis=0))  # sum along columns (year-wise total sales)
yearly_total = np.sum(sales_data[: , 1:], axis=0)
print(yearly_total)

#minimum slaes per restaurant 
print(np.min(sales_data, axis=1))  # min along rows (restaurant-wise minimum sales)
min_sales_per_restaurant = np.min(sales_data[:, 1:], axis=1)
print(min_sales_per_restaurant)
```

```python
#Maximum sales per year
max_sales_per_year = np.max(sales_data[:, 1:], axis=0)
print(max_sales_per_year)
```

```python
#Average sales per restaurant
avg_sales_per_restaurant = np.mean(sales_data[:, 1:], axis=1)
print(avg_sales_per_restaurant)
```

```python
cumsum_sales = np.cumsum(sales_data[:, 1:], axis=1)
print(cumsum_sales)

plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumsum_sales, axis=0))
plt.title("Cumulative Sales Over Years")
plt.xlabel("Years")
plt.ylabel("Cumulative Sales")
plt.grid(True)
plt.show()
```
